###create baseline model

```python
# Shuffling the Dataset
df = shuffle(df, random_state = 42)
```

```python
df.head()
```

```python
#actual value
X=df['Favorite Soft Drink']
```

```python
target_column = 'Gender'

# Extract the target values
y = df[target_column]
```

```python
# Calculate the mode of the target variable
gender_mode = y.mode()[0]
print(gender_mode)
```

```python
predictions = [gender_mode] * len(y)
```

```python
predictions
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop(target_column, axis=1), y, test_size=0.2, random_state=42)
```

```python
predictions = [gender_mode] * len(y)

accuracy = accuracy_score(y, predictions)

print(f"accuracy: {accuracy}")
```

```python
# Calculate mean values for the training set
train_classification_mode = y_train.mode()[0]
```

```python
# Create a baseline model that predicts the mean of the training set for all samples
train_predictions = [train_classification_mode] * len(y_train)
train_accuracy = accuracy_score(y_train, train_predictions)
print(f"training accuracy: {train_accuracy}")
```

```python
# Calculate precision, recall, f1-score for training
report = classification_report(y_train, train_predictions)
print(f'Gender Classification Report:\n{report}')
```

```python
df['train_classification_mode'] = train_classification_mode
```

```python
df.head()
```

```python
# Replace the y_test values with the mean of the training set
y_test_mode= pd.Series([train_classification_mode] * len(y_test), index=y_test.index)

y_test = pd.DataFrame(y_test)
y_test['train_classification_mode'] = y_test_mode
y_test
```

```python
# Calculate accuracy,precision, recall, f1-score for test dataset in training set
test_accuracy = accuracy_score(y_test['Gender'], y_test['train_classification_mode'])
report = classification_report(y_test['Gender'], y_test['train_classification_mode'])
print(f"test accuracy: {train_accuracy}")
print(f'Gender Classification Report:\n{report}')
```

##Conclusions

Since the female class is the most prevalent in this dataset, we shall provide insight based on its results.

The predictive power of the model is similar to random guessing inside the training set, as indicated by the accuracy of the 80% trained value, which is roughly 52%.
